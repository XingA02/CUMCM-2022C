```python
import pandas as pd
import matplotlib.pyplot as plt # 基础绘图库（什么都能画，但可能慢、复杂）
import seaborn as sns # 数据可视化专用库（专门用来快速画数据图）

# 忽略警告
import warnings
warnings.filterwarnings('ignore')

# 中文字体宋体，英文字体新罗马
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['font.serif'] = ['Times New Roman']
# 字体大小
plt.rcParams['font.size'] = 12  
# 正常显示负号
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('中间数据集\\表1与表2拼接后的数据集.xlsx')
df = df.drop(['纹饰', '采样位置', '颜色', '表面风化'], axis=1)
df.head()
```

```python
import pandas as pd

# 假设df是你的数据集DataFrame，category_column是分类列的名称
# 首先，找出所有不同的类别
categories = df["风化具体情况"].unique()

# 然后，为每个类别创建一个空的DataFrame
category_datasets = {category: pd.DataFrame() for category in categories}

# 遍历原始数据集，并将每个记录分配到相应的类别DataFrame中
for category in categories:
    category_datasets[category] = df[df["风化具体情况"] == category]

df1 = category_datasets["风化"]
df2 = category_datasets["无风化"]
```

```python
import pandas as pd

# 假设df是你的数据集DataFrame，category_column是分类列的名称
# 首先，找出所有不同的类别
categories = df1["类型"].unique()

# 然后，为每个类别创建一个空的DataFrame
category_datasets = {category: pd.DataFrame() for category in categories}

# 遍历原始数据集，并将每个记录分配到相应的类别DataFrame中
for category in categories:
    category_datasets[category] = df1[df1["类型"] == category]

# 风化的高钾玻璃
df_1_k = category_datasets["高钾"].drop(["类型", "风化具体情况"], axis=1)
# 风化的铅钡玻璃
df_1_b = category_datasets["铅钡"].drop(["类型", "风化具体情况"], axis=1)
```

```python
import pandas as pd

# 假设df是你的数据集DataFrame，category_column是分类列的名称
# 首先，找出所有不同的类别
categories = df2["类型"].unique()

# 然后，为每个类别创建一个空的DataFrame
category_datasets = {category: pd.DataFrame() for category in categories}

# 遍历原始数据集，并将每个记录分配到相应的类别DataFrame中
for category in categories:
    category_datasets[category] = df2[df2["类型"] == category]

# 未风化的高钾玻璃
df_2_k = category_datasets["高钾"].drop(["类型", "风化具体情况"], axis=1)
# 未风化的铅钡玻璃
df_2_b = category_datasets["铅钡"].drop(["类型", "风化具体情况"], axis=1)
```

```python
res = pd.DataFrame({
            "未风化高钾玻璃": (df_2_k.drop("文物编号", axis=1)).mean(),
            "未风化铅钡玻璃": (df_2_b.drop("文物编号", axis=1)).mean(),
            "风化高钾玻璃": (df_1_k.drop("文物编号", axis=1)).mean(),
            "风化铅钡玻璃": (df_1_b.drop("文物编号", axis=1)).mean()
            })

res["高钾玻璃风化前比风化后"] = (res["未风化高钾玻璃"]) / res["风化高钾玻璃"]
res["铅钡玻璃风化前比风化后"] = (res["未风化铅钡玻璃"]) / res["风化铅钡玻璃"]
res
```

```python
# 按比例推算高钾玻璃风化前成分
result_1 = (df_1_k.drop("文物编号", axis=1)) * res['高钾玻璃风化前比风化后']

result_1['氧化钠(Na2O)'].fillna(0.695000, inplace=True)
result_1['氧化铅(PbO)'].fillna(0.411667, inplace=True)
result_1['氧化钡(BaO)'].fillna(0.598333, inplace=True)
result_1['氧化锶(SrO)'].fillna(0.041667, inplace=True)
result_1['氧化锡(SnO2)'].fillna(0.196667, inplace=True)
result_1['二氧化硫(SO2)'].fillna(0.101667, inplace=True)

result_1["文物编号"] = df_1_k["文物编号"]
result_1.head()
```

```python
# 按比例推算铅钡玻璃风化前成分
result_2 = (df_1_b.drop("文物编号", axis=1)) * res['铅钡玻璃风化前比风化后']
result_2["文物编号"] = df_1_b["文物编号"]
result_2.head()
```

```python
# 将两个DataFrame上下合并
result_final = pd.concat([result_1, result_2], ignore_index=True)
result_final = result_final[["文物编号", "二氧化硅(SiO2)", "氧化钠(Na2O)", "氧化钾(K2O)", 
                    "氧化钙(CaO)", "氧化镁(MgO)", "氧化铝(Al2O3)", "氧化铁(Fe2O3)", 
                    "氧化铜(CuO)", "氧化铅(PbO)", "氧化钡(BaO)", "五氧化二磷(P2O5)",
                    "氧化锶(SrO)", "氧化锡(SnO2)", "二氧化硫(SO2)"]]
result_final.head()
```
